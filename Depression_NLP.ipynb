```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
file = 'depression_2019_05_25.txt'
```

### Q1. The frequency of the the "mdd"

```python
titles = []
abstracts = []
```

```python
with open(file, 'r') as f:
    for line in f:
        t_a = line.split('|') ##title and abstract is seperated by '|'
        title = t_a[0].strip() 
        abstract = t_a[1].strip()
        titles.append(title) ##append ith title to titles list
        abstracts.append(abstract) ##append ith abstract to abstracts list
```

```python
titles
```

```python
abstracts
```

```python
docs = [] ##list which stores tokens of all the titles and abstracts
for i in range(0, len(titles)):
    title = titles[i]
    abstract = abstracts[i]
    title_tokens = word_tokenize(title.lower()) #titles list elements are composed of 1 sentence
    docs.extend(title_tokens)
    sentences = sent_tokenize(abstract) #abstracts must be seperated by the sentences and then seperate by words
    for sentence in sentences:
        abstract_tokens = word_tokenize(sentence.lower())
        docs.extend(abstract_tokens)
```

```python
nltk_text = nltk.Text(docs)
```

```python
freq_dist = nltk_text.vocab()
```

```python
freq_dist['mdd']
```

### Answer
17751

## Q2. The similar words to “anxiety”

```python
nltk_text.similar('anxiety')
```

### Answer
depression the mdd and depressive mood treatment psychiatric patients symptoms stress a schizophrenia ptsd personality clinical suicide in remission pain

## Q3. “phq-9”

```python
nltk_text.concordance('phq-9')
```
